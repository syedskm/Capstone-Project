```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.support.ui import WebDriverWait
from bs4 import BeautifulSoup
import requests
```

Chennai City

```python
prod_url = []
for page in range(1, 5):
    url = f'https://www.magicbricks.com/property-for-sale/residential-real-estate?&proptype=Villas&cityName=Chennai&sortBy=mostRecent&page={page}'
    driver.get(url)
    time.sleep(5) 

    share_icons = driver.find_elements(By.XPATH, '//span[@class="mb-srp__card__share--icon"]')
    
    property_urls = []
    for icon in share_icons:
        try:
            icon.click()
            time.sleep(2) 

            share_url_elements = driver.find_elements(By.XPATH, '//div[@class="mb-srp__share-prop__flex-row--value"]')
            for url_element in share_url_elements:
                property_urls.append(url_element.text.strip())
            close_button = driver.find_element(By.XPATH, '//span[@class="mb-dialog--close"]')
            close_button.click()
            time.sleep(1) 
        except Exception as e:
            print(f"Error processing property: {e}")
            
    prod_url.extend(property_urls)
```

```python
len(prod_url)
```

```python
name = []
fea = []
price = []
price_sq = []

for page in range(1, 5):
    url = f'https://www.magicbricks.com/property-for-sale/residential-real-estate?&proptype=Villas&cityName=Chennai&sortBy=mostRecent&page={page}'
    driver.get(url)
    time.sleep(1)

    # Extracting property names
    for element in driver.find_elements(By.XPATH, '//h2[@class="mb-srp__card--title"]'):
        name.append(element.text)
    
    # Extracting property prices
    for element in driver.find_elements(By.XPATH, '//div[@class="mb-srp__card__price--amount"]'):
        price.append(element.text)
    
    # Extracting property price per sq
    for element in driver.find_elements(By.XPATH, '//div[@class="mb-srp__card__price--size"]'):
        price_sq.append(element.text)
    
    # Extracting property features
    features = []
    feature_elements = driver.find_elements(By.XPATH, '//div[@class="mb-srp__card__summary--value"]')
    for element in feature_elements:
        features.append(element.text)
    fea.append(features)
```

```python
sqft = []
status = []
floors = []

for i in fea:
    for j in i:
        if 'sqft' in j:
            sqft.append(j)
        elif "Poss." in j or 'Ready' in j:
            status.append(j)
        elif 'out of' in j:
            floors.append(j)
```

```python
def write_list_to_file(filename, data_list):
    with open(filename, 'w', encoding='utf-8') as file:
        for item in data_list:
            file.write("%s\n" % item)

# Write each list to a separate text file
write_list_to_file("sqft.txt", sqft)
write_list_to_file("status.txt", status)
write_list_to_file("floors.txt", floors)
write_list_to_file("url.txt", prod_url)
write_list_to_file("name.txt", name)
write_list_to_file("price_sq.txt", price_sq)
write_list_to_file("price.txt", price)
```

```python
with open('url.txt', 'r') as txt_file:
    url = txt_file.readlines()

url = [line.strip() for line in url]
```

```python
advance = []
for i in url:
    driver.get(i)
    time.sleep(3)
    x = []
    for i in driver.find_elements(By.XPATH,'//li[@class="mb-ldp__more-dtl__list--item"]'):
        x.append(i.text)
    advance.append(x)
```

```python
def extract_value(item, key):
    for element in item:
        if key in element:
            return element

bhk = ['Beds', 'Baths', 'Balconies', 'Parking']

beds = []
baths = []
balconies = []
parking = []

for item in bhk_fur:
    item = item.replace("', '", "', '")
    elements = item.strip("[]").split("', '")
    beds.append(extract_value(elements, 'Beds') if extract_value(elements, 'Beds') else None)
    baths.append(extract_value(elements, 'Baths') if extract_value(elements, 'Baths') else None)
    balconies.append(extract_value(elements, 'Balcony') if extract_value(elements, 'Balcony') else None)
    parking.append(extract_value(elements, 'Parking') if extract_value(elements, 'Parking') else None)

c2 = pd.DataFrame({'Beds': beds,
                   'Baths': baths,
                   'Balconies': balconies,
                   'Parking': parking})

c2 = c2.replace('None', float('nan'), regex=True)
```

```python
directions = ['North', 'South', 'East', 'West', 'North-East', 'North-West', 'South-East', 'South-West']

floors = []
transaction_type = []
house_facing = []

for item in fac_own_tra:
    floors.append(item.split(", ")[0].strip("['"))

    if 'New Property' in item:
        transaction_type.append('New Property')
    elif 'Resale' in item:
        transaction_type.append('Resale')
    else:
        transaction_type.append('Unknown')

    facing = ""
    for direction in directions:
        if direction in item:
            facing = direction
            break
    house_facing.append(facing)

c3 = pd.DataFrame({
    'Floors': floors,
    'Transaction_Type': transaction_type,
    'House_Facing': house_facing
})
```

```python
def extract_price_sqft(entry):
    entry = str(entry).strip("[]").replace("\\n", "\n").replace("â‚¹", "₹").replace(",", "")
    parts = entry.split('\n')
    
    if len(parts) > 1 and '₹' in parts[-1]:
        price_sqft = parts[-1].strip().replace('₹', '').replace('/sqft', '').replace('/sqyd', '') + '/sqft'
    else:
        price_sqft = 'unknown'
    
    return price_sqft

price_sqft_list = [extract_price_sqft(entry) for entry in ca]

c5 = pd.DataFrame(price_sqft_list, columns=['Price/Carpet Area'])

def extract_area(entry):
    entry = str(entry).strip("[]").replace("\\n", "\n").replace("â‚¹", "₹").replace(",", "")
    parts = entry.split('\n')
    
    area_value = 'Unknown'
    
    for part in parts:
        match = re.search(r'\d+', part)
        if match:
            area_value = match.group()
            break
    
    return area_value

areas = [extract_area(entry) for entry in ca]
c6 = pd.DataFrame(areas, columns=['Carpet Area'])
```

```python
def convert_to_sqft(area):
    if isinstance(area, int):
        return area
    elif isinstance(area, str):
        area = re.sub(r"[^\w\s,]", "", area)
        
        area = area.lower()
        
        if 'sqft' in area:
            return int(re.search(r'\d+', area).group())
        elif 'sqm' in area:
            sqm_value = int(re.search(r'\d+', area).group())
            return sqm_value * 10.7639  
        elif 'sqyd' in area:
            sqyd_value = int(re.search(r'\d+', area).group())
            return sqyd_value * 9 
        elif 'marla' in area:
            marla_value = int(re.search(r'\d+', area).group())
            return marla_value * 272.25  # 1 marla = 272.25 sqft
        elif 'kotha' in area:
            kotha_value = int(re.search(r'\d+', area).group())
            return kotha_value * 35000  
        return 0

df['Area'] = df['Carpet Area'].apply(convert_to_sqft)
```
